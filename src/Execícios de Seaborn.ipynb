# Exercícios

** Recrie os plots abaixo usando o DataFrame Titanic. Há poucas dicas, uma vez que a maioria dos plots pode ser feito com apenas uma ou duas linhas de código e uma dica basicamente daria a solução. Mantenha atenção aos rótulos x e y para dicas. **

** *Nota! Para não perder a imagem do plot, certifique-se de não codificar na célula que está diretamente acima do gráfico, há uma célula extra acima daquela que não substituirá esse gráfico! * **

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
g = sns.jointplot(
    x='fare',
    y='age',
    data=titanic,
    kind='scatter',
    color='steelblue'
)
g.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
plt.clf()
h = sns.histplot(
    data=titanic,
    x='fare',
    bins=30,
    color='red',
    alpha=0.4
)
h.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
plt.clf()
j = sns.boxplot(
    data=titanic,
    x='class',
    y='age',
    hue='class',
    palette=['#e67e22', '#3498db', '#2ecc71'],
    order=['First', 'Second', 'Third'],
    legend=False
)

j.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
plt.clf()

k = sns.swarmplot(
    data=titanic,
    x='class',
    y='age',
    hue='class',
    order=['First', 'Second', 'Third'],
    hue_order=['First', 'Second', 'Third'],
    palette='Set2',
    dodge=False,
    legend=False,
    size=5
)

k.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
plt.clf()

l = sns.countplot(
    data=titanic,
    x='sex',
    hue='sex',
    palette='deep',
    legend=False
)

l.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
plt.clf()
cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']
corr = titanic[cols].corr()
m = sns.heatmap(
    corr,
    cmap='coolwarm',
    vmin=-0.8,
    vmax=0.8,
    center=0
)
m.set_title('titanic.corr()')
m.figure
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.

plt.clf()
n = sns.displot(
    data=titanic,
    x='age',
    col='sex',
    bins=20,
    color='steelblue',
    height=4,
    aspect=1.2
)

n.figure
```
